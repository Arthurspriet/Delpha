```python
import folium
from folium import plugins
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
data = pd.read_csv('export.csv')
```

```python
map = folium.Map()
marker_cluster = MarkerCluster().add_to(map)
```

```python
for index, row in data.iterrows():
    folium.Marker([row["latitude"], row["longitude"]], popup=row[['Name','Headquarters','Website']],icon=folium.Icon(color='green')).add_to(marker_cluster)
```
